# VISUALIZATIONS

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# distribution (heart disease)

```python
sns.countplot(x='target', data=df)
plt.title("Heart Disease Distribution")
plt.show()
```

# age distribution

```python
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

# cholesterol distribution

```python
sns.histplot(df['chol'])
plt.title("Cholesterol distribution")
plt.show()
```

- Cholesterol values are slightly right-skewed, with most values between 200–300..

# chest pain vs heart disease

```python
sns.countplot(x='cp', hue='target', data=df)
plt.title("Chest Pain vs Heart Disease")
plt.show()
```

# sex vs heart disease

```python
sns.countplot(x='sex', hue='target', data=df)
plt.title("Gender vs heart disease")
plt.show()
```


**Males have higher heart disease cases compared to females.

**Gender plays an important role in heart disease occurrence.


# Target vs Chest Pain 

```python
sns.countplot(x='cp', hue='target', data=df)
plt.title("Chest Pain vs Heart Disease")
plt.show()
```

# Insights

- Most patients fall in age range 40–60.
- Higher cholesterol is linked with heart disease.
- Chest pain type 2 and 3 are more common in heart patients.
- Exercise-induced angina is present in ~33% patients.
- Males have higher heart disease cases than females.

# Conclusion

This analysis highlights key factors associated with heart disease:

- Higher cholesterol levels are commonly observed in patients.
- Blood pressure tends to be elevated among affected individuals.
- Exercise-induced angina is present in a significant portion of patients (~33%).
- Males show a higher likelihood of heart disease compared to females.

These findings can help in early risk identification and preventive healthcare planning.

This analysis can assist healthcare professionals in identifying high-risk patients early.
